# Concatenation in pandas

```python
#create two dataframe
df1=pd.DataFrame({
    "A":[1,2,3,4,5],
    "B":[4,3,2,1,9],
    "C":[34,32,45,56,67]
}
    )
df1
```

```python
df2=pd.DataFrame({
    "D":[4,6,8,8,10],
    "E":[43,54,67,32,11],
}
    )
df2
```

```python
#concat()
df3=pd.concat([df1,df2],axis=1) #one vaneko row wise concatination #1 vaneko column wise concatination naya features add garna 
df3
```

# Handling Missing Value

```python
#np.nan vaneko null value
student=["student1","student2","student1","student4","student5","student6","student7"]
age=[21,23,21,19,np.nan,20,np.nan]
gender=["Male","Female","Male","Male",np.nan,"Female","Male"]
marks=[45,54,45,64,71,np.nan,56]
df_students=pd.DataFrame({"students":student,"Age":age,"Gender":gender,"marks_obtained":marks})
df_students
```

```python
#To remove duplicate data
df_students.duplicated().sum() #.sum() function hale pachi kati ota duplicate data cha vanera dhekhaucha
```

```python
df_students.drop_duplicates(inplace=True)
df_students #tala call garda duplicate hateko hudaina
```

```python
df_students.isnull().sum() #to check the null value c
```

# Techniques to handle null/missing values

### first method :remove the row that contains missing value

```python
#dropna()
df_students.dropna()
```

### second method :forward filling and backward filling

```python
df_students
```

```python
# fillna()
df_students.fillna(method="ffill")
```

```python
df_students.fillna(method="bfill") #backward fill b
```

# method 3 :mean,median and mode imputation

```python
mean_age=df_students['Age'].mean()
mean_age
```

```python
mean_marks=df_students["marks_obtained"].mean()
mean_marks
```

```python
gender_mode=df_students["Gender"].mode()[0]
gender_mode
```

```python
df_filled=df_students.copy()
df_filled
```

```python
df_filled["Age"]=df_filled["Age"].fillna(mean_age)
df_filled
```

```python
df_filled["marks_obtained"]=df_filled["marks_obtained"].fillna(mean_marks)
df_filled
```

```python
df_filled["Gender"]=df_filled["Gender"].fillna(gender_mode)
df_filled
```
