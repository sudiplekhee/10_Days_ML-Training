# pie chart

```python
x=["TV","Radio","Newspaper","Posters"]
y=[60,25,15,10]
explode=[0.3,0.2,0.6,0.4]
plt.pie(y,labels=x,autopct="%.1f%%",shadow=True,explode=explode)   #only numerical value are allowed
plt.show()
```
