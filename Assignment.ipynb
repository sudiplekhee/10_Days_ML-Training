## splitting x and y

```python
df.columns
```

```python
x = df.drop(columns='label')
```

```python
x.head()
```

```python
y = df['label']
```

```python
y.head()
```

## spliting the x and y training dataset

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## model creation

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
crop_model = lr.fit(x_train, y_train)
```

```python
crop_model
```

```python
predictions1= crop_model.predict(x_test)
```

```python
predictions1.shape
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print(f"Accuracy Score:{accuracy_score(y_test,predictions1)}")
```

```python
result1 = pd.DataFrame({
    "Actual": y_test,
    "Predicted" : predictions1
})
```

```python
result1.head()
```

```python
plt.scatter(y_test, predictions1)
plt.show()
```

```python
sns.heatmap(confusion_matrix(y_test, predictions1), annot=True)
plt.xlabel("Predicated")
plt.ylabel("Actual")
plt.show()
```
